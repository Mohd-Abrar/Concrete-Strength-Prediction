### Performing hypothesis testing to find the significant variables

Hypothesis :

H0 (Null Hypothesis) : There is no significant relationship between the variables being tested.

Ha (Alternative Hypothesis) : There is a significant relationship between the variables being tested

Consider significance level as 0.05

```python
from scipy import stats
```

```python
num_cols = df.select_dtypes(include = np.number).columns.to_list()

num_cols.remove('compressive_strength')

print(num_cols)
```

```python
# Num vs Num - f_oneway test

statistical_result = pd.DataFrame(columns = ['Columns','Pvalue','Remarks'])

for i in num_cols:
    
    stat , pval = stats.pearsonr(df[i] , df['compressive_strength'])
    
    statistical_result = statistical_result.append({'Columns': i,
                                                   'Pvalue':pval,
                                                   'Remarks':'Reject H0' if pval <= 0.05 else 'Failed to reject H0'},
                                                  ignore_index = True)
    
statistical_result
```

#### Insights from statistical test

From above performed statistical tests we can conclude that all of the columns has rejected null hypotheses which means the columns are significant to the target variable compressive_strength.

### Splitting the dataset randomly into train and test dataset using ratio of 70:30

```python
from sklearn.model_selection import train_test_split
```

```python
x = df.drop(columns = 'compressive_strength')
y = df['compressive_strength']

xtrain , xtest , ytrain , ytest = train_test_split(x , y , test_size = 0.30)
```

### Checking and treating of outliers

```python
f , ax = plt.subplots(3,3)

for i , v in zip(df.columns , ax.flatten()):
    sns.boxplot(y = df[i] , ax = v)
    
plt.tight_layout()
plt.show()
```

From above boxplot it is clearly evident that there outliers present in water , super_plasticizer , age , fine_aggregate. By doing IQR method we tend to lose data so we go forward by performing PowerTransformer technique

```python
from sklearn.preprocessing import PowerTransformer
```

```python
out_cols = ['water' , 'super_plasticizer' , 'age' ,'fine_aggregate']

pt = PowerTransformer(standardize=False)

for i in out_cols :
    variable = pt.fit(xtrain[[i]])
    
    xtrain[i] = variable.transform(xtrain[[i]])
    xtest[i] = variable.transform(xtest[[i]])
```

```python
xtrain
```

### Building a base model

Building a base model using Linear Regression as it is having the highest explanatory power compared to other models

```python
import statsmodels.api as sma
```

```python
model_lr = sma.OLS(ytrain , sma.add_constant(xtrain)).fit()

model_lr
```

#### Checking for summary

```python
model_lr.summary()
```

```python
from sklearn.metrics import mean_squared_error , r2_score

pred_train = model_lr.predict(sma.add_constant(xtrain))
pred_test = model_lr.predict(sma.add_constant(xtest))

rmse_train = np.sqrt(mean_squared_error(ytrain , pred_train))
rmse_test = np.sqrt(mean_squared_error(ytest , pred_test))

print('RMSE Train',rmse_train)
print('RMSE Test',rmse_test)
```

```python
# Creating a Dataframe to store metrics of train and test

performance_df = pd.DataFrame(columns = ['Model','R2_train','R2_test','RMSE_train','RMSE_test','Remarks'])

performance_df
```

```python
# Appending values of base model in dataframe

performance_df = performance_df.append({'Model':'Linear Regression',
                                       'R2_train': r2_score(ytrain,pred_train),
                                       'R2_test': r2_score(ytest,pred_test),
                                       'RMSE_train':rmse_train,
                                       'RMSE_test':rmse_test,
                                       'Remarks' : 'Base'} , ignore_index=True)

performance_df
```

### Building different models and evaluating using appropriate technique

```python
# Creating a user defined function to predict and to store metrics of train and test


def model_performance(model,name):
    global performance_df
    
    pred_train = model.predict(xtrain)
    pred_test = model.predict(xtest)
    
    rmse_train = np.sqrt(mean_squared_error(ytrain,pred_train))
    rmse_test = np.sqrt(mean_squared_error(ytest,pred_test))
    r2_train = r2_score(ytrain , pred_train)
    r2_test = r2_score(ytest , pred_test)
    
    def remark(r2_train,r2_test,rmse_test):
        if abs(r2_train - r2_test) > 0.1 or rmse_test > 7.5:
            return 'Over Fit'
        
        else:
            return 'Good Fit'
        
    performance_df = performance_df.append({'Model':name,
                                           'R2_train':r2_train,
                                           'R2_test':r2_test,
                                           'RMSE_train':rmse_train,
                                           'RMSE_test':rmse_test,
                                           'Remarks':remark(r2_train,r2_test,rmse_test)} , ignore_index=True)
```

```python
# Decision tree 

from sklearn.tree import DecisionTreeRegressor

model_dt = DecisionTreeRegressor().fit(xtrain , ytrain)

model_performance(model_dt,'DecisionTree')
```

```python
# RandomForest

from sklearn.ensemble import RandomForestRegressor

model_rf = RandomForestRegressor().fit(xtrain,ytrain)

model_performance(model_rf , 'Random Forest')
```

```python
# KNN

from sklearn.neighbors import KNeighborsRegressor

model_knn = KNeighborsRegressor().fit(xtrain , ytrain)

model_performance(model_knn , 'KNN')
```

```python
# AdaBoost

from sklearn.ensemble import AdaBoostRegressor

model_ab = AdaBoostRegressor().fit(xtrain , ytrain)

model_performance(model_ab , 'AdaBoost')
```

```python
# Gradient Boosting

from sklearn.ensemble import GradientBoostingRegressor

model_gb = GradientBoostingRegressor().fit(xtrain , ytrain)

model_performance(model_gb , 'Gradient Boosting')
```

```python
# Xgboost-RandomForest

from xgboost import XGBRFRegressor

model_xgbrf = XGBRFRegressor().fit(xtrain,ytrain)

model_performance(model_xgbrf , 'XGBoost Random Forest')
```

```python
# Neural network

from sklearn.neural_network import MLPRegressor

model_nn = MLPRegressor().fit(xtrain , ytrain)

model_performance(model_nn , 'Neural Network')
```

```python
# Creating a user defined function to highlight the rows which are good fit

def highlight_row(df):
    color_green = ['background-color : lightgreen']*len(df)
    color_white = ['background-color : white']*len(df)
    
    if df['Remarks'] == 'Good Fit':
        return color_green
    
    else:
        return color_white
```

```python
performance_df.style.apply(highlight_row,axis = 1)
```

After assessing various models, it was observed that some models exhibited a significant drop in performance when applied to unseen data, indicating overfitting. However, there were models that consistently performed well on both training and unseen data. Notably, the Gradient boosting model outperformed other models in terms of high r2 and low rmse. Hence, based on its superior performance and generalization ability, we can confidently consider the Gradient boosting model as our final choice.
